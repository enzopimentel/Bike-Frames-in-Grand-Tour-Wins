### Import libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### Read Website

```python
URL_2022_TDF_GC = 'https://firstcycling.com/race.php?r=17&y=2022'

r = requests.get(URL_2022_TDF_GC)

soup = BeautifulSoup(r.content,'html.parser')

print(soup.prettify())
```

```python
tdf_table_gc = soup.find('div', class_='tab-content results').find_all('tr')
tdf_table_gc
```

### 2022 TDF
